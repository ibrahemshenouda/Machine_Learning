```python
#Libraries import
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import median_absolute_error
import numpy as np
import pandas as pd
```

```python
#read data
df = pd.read_csv('houses.csv')
df.head(5)
```

```python
#clean data and seprate features and target
df_cleaned = df.dropna()
X = df_cleaned.drop('price', axis=1)
y = df_cleaned['price']

print(X.shape)
print(y.shape)
```

```python
#split data to train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44, shuffle=True)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#apply minmaxscaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
#fit and transform for train data
X_train_scaled = scaler.fit_transform(X_train)
#only transform test data, to prevent data leakage
X_test_scaled = scaler.transform(X_test)

print("Shape of scaled X_train:", X_train_scaled.shape)
print("Shape of scaled X_test:", X_test_scaled.shape)
```

```python
#fit linear regression model
LRM = LinearRegression(fit_intercept=True, copy_X=True, n_jobs=-1)
LRM.fit(X_train, y_train)
```

```python
#print score for train and test
print(LRM.score(X_train, y_train))
print(LRM.score(X_test, y_test))
```

```python
#calculate y_pred to calculate metrics
y_pred = LRM.predict(X_test)
print(f"Mean Squared Error (MSE): {mean_squared_error(y_test, y_pred):.2f}")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test, y_pred):.2f}")
print(f"Median Absolute Error (MedAE): {median_absolute_error(y_test, y_pred):.2f}")
print(f"R-squared score: {LRM.score(X_test, y_test):.2f}")
```
