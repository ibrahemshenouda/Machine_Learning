```python
from sklearn.datasets import load_iris
IrisData = load_iris()

x = IrisData.data

print(x[:3])
print(x.shape)
print(IrisData.feature_names)

y = IrisData.target
print(y[:150])
print(y.shape)
print(IrisData.target_names)
```

```python
from sklearn.datasets import load_digits
DigitsData = load_digits()

x = DigitsData.data

print(x[:3])
print(x.shape)
print(DigitsData.feature_names)
y = DigitsData.target

print(y[:100])
print(y.shape)
print(DigitsData.target_names)


import matplotlib.pyplot as plt
plt.gray()

for g in range(2):
  print(g)
  plt.matshow(DigitsData.images[g])
  plt.show()

```

```python
from sklearn.datasets import load_wine
WineData = load_wine()

x = WineData.data

print(x[:3])
print(x.shape)
print(WineData.feature_names)
y = WineData.target

print(y[:178])
print(y.shape)
```

```python
from sklearn.datasets import load_breast_cancer
CancerData = load_breast_cancer()

x = CancerData.data

print(x[:3])
print(x.shape)
print(CancerData.feature_names)
y = CancerData.target

print(y[:3])
```

```python
from sklearn.datasets import make_regression

x, y = make_regression(n_samples=20,
                       n_features= 3, shuffle=True)

print(x[:10])
print(x.shape)

print(y[:10])
print(y.shape)
```

```python
from sklearn.datasets import make_classification

x, y = make_classification(n_samples=20,
                           n_classes = 2,
                           n_features= 3,
                           n_informative=1,
                           n_clusters_per_class=1,
                           shuffle=True)

print(x[:10])
print(x.shape)

print(y[:10])
print(y.shape)
```
