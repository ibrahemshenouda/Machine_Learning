**A.Regression Metrics**

1.Mean Abslute Error

```python
from sklearn.metrics import mean_absolute_error

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2 ,8]
mae = mean_absolute_error(y_true, y_pred)
print("Mean Absolute Error_1:", mae)


y_true = [[0.5, 1], [-1, 1], [7, -6]]
y_pred = [[0, 2], [-1, 2], [8, -5]]
# uniform_average is the default method if you dont change, you can use raw_values
mae = mean_absolute_error(y_true, y_pred,multioutput='uniform_average')
print("Mean Absolute Error_2:", mae)

mae = mean_absolute_error(y_true, y_pred,multioutput='raw_values')
print("Mean Absolute Error_3:", mae)
```

2.Mean Square Error

```python
from sklearn.metrics import mean_squared_error


y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2 ,8]
mse = mean_squared_error(y_true, y_pred)
print("Mean Squared Error_1:", mse)

y_true = [[0.5, 1], [-1, 1], [7, -6]]
y_pred = [[0, 2], [-1, 2], [8, -5]]
mse = mean_squared_error(y_true, y_pred,multioutput='uniform_average')
print("Mean Squared Error_2:", mse)

mse = mean_squared_error(y_true, y_pred,multioutput='raw_values')
print("Mean Squared Error_3:", mse)
```

3.Median Absolute Error

```python
from sklearn.metrics import median_absolute_error


y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2 ,8]
mdae = median_absolute_error(y_true, y_pred)
print("Mean Squared Error_1:", mdae)

y_true = [[0.5, 1], [-1, 1], [7, -6]]
y_pred = [[0, 2], [-1, 2], [8, -5]]
mdae = median_absolute_error(y_true, y_pred,multioutput='uniform_average')
print("Mean Squared Error_2:", mdae)

mdae = median_absolute_error(y_true, y_pred,multioutput='raw_values')
print("Mean Squared Error_3:", mdae)
```
