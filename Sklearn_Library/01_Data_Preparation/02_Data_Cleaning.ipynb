```python
from sklearn.impute import SimpleImputer
import numpy as np


data = [[1,2,np.nan],
        [3,np.nan,1],
        [5,np.nan,0],
        [np.nan,4,6 ],
        [5,0,np.nan],
        [4,5,5]]
#missing_values ex: 0, nan, ...
#you can use moean or median
imp = SimpleImputer(missing_values=np.nan,
                    strategy='mean')
imp.fit(data)

modified_data = imp.transform(data)
print(modified_data)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np


data = [[1,2,0],
        [3,0,1],
        [5,0,0],
        [0,4,6],
        [5,0,0],
        [4,5,5]]

imp = SimpleImputer(missing_values=0,
                    strategy='mean')
imp.fit(data)

modified_data = imp.transform(data)
print(modified_data)
```
