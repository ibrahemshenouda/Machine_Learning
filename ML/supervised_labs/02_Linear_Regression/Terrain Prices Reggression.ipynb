### Load Data
First, we load the `train.csv` and `test.csv` files into pandas DataFrames.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

df_train = pd.read_csv('/content/train.csv')
df_test = pd.read_csv('/content/test.csv')

print("Train data head:")
display(df_train.head())

print("Test data head:")
display(df_test.head())
```

### Dropping the 'id' Column
We will drop the 'id' column from both the training and testing datasets as it is typically not used for model training. We'll keep copies of the original dataframes to retain the 'id' for potential submission later.

```python
df_train_original = df_train.copy() # Keep original for submission id
df_test_original = df_test.copy() # Keep original for submission id

if 'id' in df_train.columns:
    df_train = df_train.drop('id', axis=1)
if 'id' in df_test.columns:
    df_test = df_test.drop('id', axis=1)

print("df_train head after dropping 'id' column:")
display(df_train.head())

print("df_test head after dropping 'id' column:")
display(df_test.head())
```

### Handling Categorical Features (Label Encoding)
Categorical features need to be converted into numerical representations for machine learning models. We will use Label Encoding for this purpose. The encoder will be fitted on the training data and then used to transform both training and test data. For unseen categories in the test set, we will assign a value of -1.

```python
# Identify non-numerical columns
non_numeric_cols_train = df_train.select_dtypes(include=['object', 'category']).columns
non_numeric_cols_test = df_test.select_dtypes(include=['object', 'category']).columns

# Combine all unique non-numeric columns from both dataframes
all_non_numeric_cols = list(set(list(non_numeric_cols_train) + list(non_numeric_cols_test)))

print(f"Non-numerical columns to encode: {all_non_numeric_cols}")

# Apply Label Encoding
for col in all_non_numeric_cols:
    le = LabelEncoder()
    if col in df_train.columns:
        # Fit on training data and transform
        df_train[col] = le.fit_transform(df_train[col])
        print(f"Encoded '{col}' in df_train: unique values before: {len(le.classes_)}")
    if col in df_test.columns:
        train_labels = list(le.classes_)
        mapping = {label: idx for idx, label in enumerate(train_labels)}

        def apply_label_encoding(series, mapping):
            return series.map(mapping).fillna(-1) # Use -1 for unseen categories

        df_test[col] = apply_label_encoding(df_test[col], mapping)
        df_test[col] = df_test[col].astype(int) # Convert back to int after fillna for consistency
        print(f"Encoded '{col}' in df_test using df_train's encoder.")


print("\n--- df_train head after encoding ---")
display(df_train.head())
print("\n--- df_test head after encoding ---")
display(df_test.head())

print("\n--- df_train info after encoding ---")
df_train.info()
print("\n--- df_test info after encoding ---")
df_test.info()
```

### Handling Outliers using IQR Method
To manage outliers, I'll implement a capping strategy based on the Interquartile Range (IQR). This involves defining lower and upper bounds (Q1 - 1.5 * IQR and Q3 + 1.5 * IQR, respectively) and then replacing any values outside these bounds with the respective boundary value. This approach helps to mitigate the impact of extreme values while retaining all data points.

```python
def cap_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap the values
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df

# Identify numerical columns (excluding the target in df_train)
numerical_cols_train = df_train.select_dtypes(include=np.number).columns.tolist()
if 'target' in numerical_cols_train:
    numerical_cols_train.remove('target')
numerical_cols_test = df_test.select_dtypes(include=np.number).columns.tolist()

print("Applying outlier capping to numerical columns in df_train:")
for col in numerical_cols_train:
    df_train = cap_outliers_iqr(df_train, col)
    print(f"  - Capped outliers for: {col}")

print("\nApplying outlier capping to numerical columns in df_test:")
for col in numerical_cols_test:
    df_test = cap_outliers_iqr(df_test, col)
    print(f"  - Capped outliers for: {col}")

print("\n--- df_train descriptive statistics after outlier capping ---")
display(df_train.describe())

print("\n--- df_test descriptive statistics after outlier capping ---")
display(df_test.describe())
```

### Feature and Target Scaling using MinMaxScaler
We will now apply Min-Max Scaling to all numerical features in both the training and test datasets, as well as to the target variable in the training set. This transformation scales features to a given range (typically 0 to 1), which can be beneficial for many machine learning algorithms. The scaler will be fitted on the training data and then used to transform both training and test data to prevent data leakage.

```python
# Separate features (X) and target (y) from the training data
X = df_train.drop('target', axis=1)
y = df_train['target']

# Initialize MinMaxScaler for features
feature_scaler = MinMaxScaler()

# Fit the scaler on training features and transform them
X_train_scaled = feature_scaler.fit_transform(X)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X.columns)

# Transform the test features using the same scaler fitted on the training data
df_test_scaled = feature_scaler.transform(df_test)
df_test_scaled = pd.DataFrame(df_test_scaled, columns=df_test.columns)

# Assign X_train to the scaled features
X_train = X_train_scaled.copy()

print("\n--- X_train (scaled features) head ---")
display(X_train.head())

print("\n--- df_test (scaled features) head ---")
display(df_test_scaled.head())
```

```python
# Initialize MinMaxScaler for the target variable
target_scaler = MinMaxScaler()

# Fit the scaler on the training target and transform it
y_train_scaled = target_scaler.fit_transform(y.values.reshape(-1, 1))
y_train_scaled = pd.Series(y_train_scaled.flatten(), name='target_scaled')

print("\n--- y_train_scaled head ---")
display(y_train_scaled.head())
```

### Prepare Features and Target for Modeling
We now define `X_train` (features for training), `y_train_scaled` (scaled target for training), and `X_test` (features for generating predictions).

```python
# X_train is already defined and scaled from the previous step
X_train = X_train # Renaming for clarity as X_train already contains features from df_train

# X_test is the entire df_test as it does not contain a target column
X_test = df_test.copy()

# y_test is not available for the test dataset
y_test = None

print("\n--- X_train head after separation (scaled features) ---")
display(X_train.head())

print("\n--- X_test head after separation (scaled features) ---")
display(X_test.head())

print("\n--- y_train_scaled head after separation ---")
display(y_train_scaled.head())
```

### Split Training Data for Validation
To evaluate model performance objectively, we split the preprocessed training data (`X_train`, `y_train_scaled`) into an 80/20 training/validation set. This allows us to train models on `X_train_new` and `y_train_new_scaled` and evaluate them on `X_test_new` and `y_test_new_scaled`.

```python
# Split the scaled training data into training and a new 'test' (validation) set
X_train_new, X_test_new, y_train_new_scaled, y_test_new_scaled = train_test_split(
    X_train, y_train_scaled, test_size=0.2, random_state=42
)

print("Shape of X_train_new:", X_train_new.shape)
print("Shape of X_test_new:", X_test_new.shape)
print("Shape of y_train_new_scaled:", y_train_new_scaled.shape)
print("Shape of y_test_new_scaled:", y_test_new_scaled.shape)
```

### Linear Regression Model Training and Evaluation
We will start by training a simple Linear Regression model as a baseline. After training, we will evaluate its performance using metrics such as R-squared, Mean Squared Error (MSE), and Mean Absolute Error (MAE) on both the training and validation sets.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Initialize and train the Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train_new, y_train_new_scaled)

# Make predictions on the training and validation sets
y_train_pred_new_scaled = linear_model.predict(X_train_new)
y_test_pred_new_scaled = linear_model.predict(X_test_new)

# Evaluate the model on the training set
r2_train_new_scaled = r2_score(y_train_new_scaled, y_train_pred_new_scaled)
mse_train_new_scaled = mean_squared_error(y_train_new_scaled, y_train_pred_new_scaled)
mae_train_new_scaled = mean_absolute_error(y_train_new_scaled, y_train_pred_new_scaled)
rmse_train_new_scaled = np.sqrt(mse_train_new_scaled)

# Adjusted R-squared for training set
n_train_new = X_train_new.shape[0]
p_train_new = X_train_new.shape[1]
adjusted_r2_train_new_scaled = 1 - (1 - r2_train_new_scaled) * (n_train_new - 1) / (n_train_new - p_train_new - 1)

# Evaluate the model on the validation set
r2_test_new_scaled = r2_score(y_test_new_scaled, y_test_pred_new_scaled)
mse_test_new_scaled = mean_squared_error(y_test_new_scaled, y_test_pred_new_scaled)
mae_test_new_scaled = mean_absolute_error(y_test_new_scaled, y_test_pred_new_scaled)
rmse_test_new_scaled = np.sqrt(mse_test_new_scaled)

# Adjusted R-squared for validation set
n_test_new = X_test_new.shape[0]
p_test_new = X_test_new.shape[1]
adjusted_r2_test_new_scaled = 1 - (1 - r2_test_new_scaled) * (n_test_new - 1) / (n_test_new - p_test_new - 1)

print("\nLinear Regression Model Performance (Scaled Data):\n")
print(f"Training R-squared: {r2_train_new_scaled:.4f}")
print(f"Training Adjusted R-squared: {adjusted_r2_train_new_scaled:.4f}")
print(f"Training MSE: {mse_train_new_scaled:.4f}")
print(f"Training RMSE: {rmse_train_new_scaled:.4f}")
print(f"Training MAE: {mae_train_new_scaled:.4f}")

print(f"\nValidation R-squared: {r2_test_new_scaled:.4f}")
print(f"Validation Adjusted R-squared: {adjusted_r2_test_new_scaled:.4f}")
print(f"Validation MSE: {mse_test_new_scaled:.4f}")
print(f"Validation RMSE: {rmse_test_new_scaled:.4f}")
print(f"Validation MAE: {mae_test_new_scaled:.4f}")
```

### Hyperparameter Tuning for Ridge Regression using GridSearchCV
Given the promising performance of the Ridge Regression model, we will now perform hyperparameter tuning to find its optimal `alpha` value. We will use `GridSearchCV` to systematically search through a range of `alpha` values and identify the one that yields the best performance on our validation set. This helps prevent overfitting and ensures the model generalizes well to unseen data.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

# Define the parameter grid for alpha
param_grid = {'alpha': np.logspace(-4, 2, 100)}

# Initialize GridSearchCV with Ridge regressor
# Using neg_mean_squared_error as scoring to optimize for lower MSE
grid_search_ridge = GridSearchCV(Ridge(random_state=42), param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

# Fit GridSearchCV to the training data
grid_search_ridge.fit(X_train_new, y_train_new_scaled)

# Get the best alpha parameter
best_alpha_ridge = grid_search_ridge.best_params_['alpha']
print(f"Best alpha for Ridge Regression: {best_alpha_ridge:.4f}")

# Train the Ridge model with the best alpha
best_ridge_model = Ridge(alpha=best_alpha_ridge, random_state=42)
best_ridge_model.fit(X_train_new, y_train_new_scaled)

# Make predictions on the training and validation sets with the best model
y_train_pred_tuned_ridge_scaled = best_ridge_model.predict(X_train_new)
y_test_pred_tuned_ridge_scaled = best_ridge_model.predict(X_test_new)

# Evaluate the tuned model on the training set
r2_train_tuned_ridge_scaled = r2_score(y_train_new_scaled, y_train_pred_tuned_ridge_scaled)
mse_train_tuned_ridge_scaled = mean_squared_error(y_train_new_scaled, y_train_pred_tuned_ridge_scaled)
mae_train_tuned_ridge_scaled = mean_absolute_error(y_train_new_scaled, y_train_pred_tuned_ridge_scaled)
rmse_train_tuned_ridge_scaled = np.sqrt(mse_train_tuned_ridge_scaled)
adjusted_r2_train_tuned_ridge_scaled = 1 - (1 - r2_train_tuned_ridge_scaled) * (n_train_new - 1) / (n_train_new - p_train_new - 1)

# Evaluate the tuned model on the validation set
r2_test_tuned_ridge_scaled = r2_score(y_test_new_scaled, y_test_pred_tuned_ridge_scaled)
mse_test_tuned_ridge_scaled = mean_squared_error(y_test_new_scaled, y_test_pred_tuned_ridge_scaled)
mae_test_tuned_ridge_scaled = mean_absolute_error(y_test_new_scaled, y_test_pred_tuned_ridge_scaled)
rmse_test_tuned_ridge_scaled = np.sqrt(mse_test_tuned_ridge_scaled)
adjusted_r2_test_tuned_ridge_scaled = 1 - (1 - r2_test_tuned_ridge_scaled) * (n_test_new - 1) / (n_test_new - p_test_new - 1)

print("\nOptimized Ridge Regression Model Performance (Scaled Data):\n")
print(f"Training R-squared: {r2_train_tuned_ridge_scaled:.4f}")
print(f"Training Adjusted R-squared: {adjusted_r2_train_tuned_ridge_scaled:.4f}")
print(f"Training MSE: {mse_train_tuned_ridge_scaled:.4f}")
print(f"Training RMSE: {rmse_train_tuned_ridge_scaled:.4f}")
print(f"Training MAE: {mae_train_tuned_ridge_scaled:.4f}")

print(f"\nValidation R-squared: {r2_test_tuned_ridge_scaled:.4f}")
print(f"Validation Adjusted R-squared: {adjusted_r2_test_tuned_ridge_scaled:.4f}")
print(f"Validation MSE: {mse_test_tuned_ridge_scaled:.4f}")
print(f"Validation RMSE: {rmse_test_tuned_ridge_scaled:.4f}")
print(f"Validation MAE: {mae_test_tuned_ridge_scaled:.4f}")
```
